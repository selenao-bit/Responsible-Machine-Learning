## Governance Memo: Model Behavior, Explanation Findings, and Monitoring Recommendations

**To:** Court Audit and Oversight Division  
**From:** Data Analyst, Risk Assessment Review Team  
**Subject:** Governance Review of COMPAS Replacement Model  
**Date:** 03/30/2003

This memo summarizes the behavior of the Gradient-Boosted Tree (GBT) recidivism model and evaluates its transparency using SHAP, LIME, and DiCE explanation methods. The goal is to assess whether the model’s decision-making aligns with principles of fairness, interpretability, and responsible use in judicial contexts.

Across all explanation techniques, the model consistently relies on a small set of features: prior criminal history (`priors_count`), COMPAS score category (`score_factor`), age group (`age_factor`), and charge severity (`crime_factor`). SHAP’s global analysis shows that higher prior counts and a “High” COMPAS score exert the strongest upward pressure on predicted risk. LIME’s local explanations confirm this pattern for individual defendants, revealing that these same features dominate case-level predictions. DiCE counterfactuals further demonstrate that substantial increases in prior counts or shifts in age category are typically required to flip an individual’s predicted outcome.

While race is not used directly as a model input, group-level fairness metrics reveal persistent disparities. African-American defendants experience higher false positive rates, while Caucasian defendants experience higher false negative rates. These patterns mirror known issues in risk assessment tools and suggest that correlated features—particularly prior history—may propagate structural inequities.

Each explanation method has limitations. SHAP can obscure causal pathways due to feature correlations; LIME’s local approximations may be unstable; and DiCE counterfactuals depend heavily on the feasibility of proposed changes. None of these methods alone provides a complete picture.

I recommend ongoing monitoring that includes regular fairness audits across demographic groups, routine checks to ensure that explanations remain stable over time, careful evaluation of alternative feature sets that reduce reliance on historically biased variables, and continued human oversight for high‑stakes decisions. Together, these practices will help ensure that the model remains transparent, accountable, and aligned with judicial fairness standards.
