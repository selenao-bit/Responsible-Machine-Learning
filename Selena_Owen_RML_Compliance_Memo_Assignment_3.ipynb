## **Compliance Memo: Fairness Assessment of COMPAS‑Style Risk Model**

**To:** Office of Algorithmic Oversight and Regulatory Compliance  
**From:** Data Analyst, Model Risk & Fairness Review Unit  
**Subject:** Fairness Evaluation of Recidivism Risk Classification Model  
**Date:** 04/06/2026

This memo summarizes the results of a fairness assessment conducted on a COMPAS‑style recidivism prediction model using demographic disparity metrics and error‑rate analysis. The evaluation examined whether the model’s high‑risk classifications are distributed equitably across racial and sex‑based groups and whether misclassification patterns indicate potential disparate impact.

We first computed three standard group‑fairness metrics—Adverse Impact Ratio (AIR), Mean Difference (ME), and Standardized Mean Difference (SMD)—for race and sex. These metrics consistently showed that African‑American and Native American defendants experience substantially higher high‑risk classification rates relative to Caucasian defendants, the reference group. For example, Native American individuals exhibited an AIR exceeding 3.4, indicating they are labeled high‑risk more than three times as often as the reference group. Sex‑based disparities were smaller but still present, with females receiving high‑risk labels at approximately 71% the rate of males. Intersectional analysis revealed even stronger disparities: Native American women had the highest AIR (≈6.3), demonstrating compounded disadvantage not visible in single‑attribute analyses.

Error‑rate analysis further highlighted systematic imbalances. African‑American defendants had a significantly higher false positive rate (0.228 vs. 0.083 for Caucasians), while Caucasian defendants had a higher false negative rate (0.720 vs. 0.492). Two‑proportion z‑tests confirmed both disparities as statistically significant (p < 0.001), indicating that these differences are unlikely to be due to sampling variability.

Limitations include potential unobserved confounders, historical bias embedded in prior‑record variables, and small sample sizes for certain subgroups (e.g., Native American and Asian defendants), which may inflate variance. Nonetheless, the consistency of disparities across multiple metrics suggests that the model’s outputs warrant continued monitoring, feature‑level review, and consideration of mitigation strategies to ensure compliance with fairness and nondiscrimination standards.
